```python
nasdaq_screener = pd.read_csv("nasdaq_screener.csv")
ticker_symbol_list = list(nasdaq_screener["Symbol"])
ticker_symbol_list
```

```python
data = yf.Ticker("MSFT")
df = data.history(period="1y")
df.index
```

```python
not_found_symbol = []
span = "1y"
for symbol in ticker_symbol_list:
    print(f"Processing: {symbol}")
    try:
        ticker = yf.Ticker(symbol)
        df = ticker.history(period=span)
        
        if df.empty:
            print(f"No data found for {symbol}.")
            not_found_symbol.append(symbol)
        else:
            df.to_csv(f"data/{symbol}_{span}.csv")
            print(f"Data saved for {symbol}.")
    
    except Exception as e:
        print(f"Error processing {symbol}: {e}")
        not_found_symbol.append(symbol)
```
